```python
import  matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
Auto = pd.read_csv("https://raw.githubusercontent.com/Riuchek/Python/main/IA/Aula%2002/Auto.csv")
Auto
```

```python
Auto_New = pd.read_csv('https://raw.githubusercontent.com/Riuchek/Python/main/IA/Aula%2002/Auto.csv', na_values=['?'])
Auto_New=Auto.dropna()
Auto_New
```

```python
Auto_New
Auto_New.nunique()
```

```python
Auto = Auto_New
```

```python
Auto.mpg.min(), Auto.mpg.max()
```

```python
for c in Auto.columns:
    print(c, "  ", Auto[c].min(), ", ", Auto[c].max())
```

```python
Auto.hist();
```

```python
fig, eixo = plt.subplots(nrows=2,ncols=3,figsize=(15,7))
Auto.plot.scatter('cylinders','mpg',ax=eixo[0,0])
Auto.plot.scatter('displacement','mpg',ax=eixo[0,1])
Auto.plot.scatter('horsepower','mpg',ax=eixo[0,2])
Auto.plot.scatter('weight','mpg',ax=eixo[1,0])
Auto.plot.scatter('acceleration','mpg',ax=eixo[1,1])
Auto.plot.scatter('year','mpg',ax=eixo[1,2])
```

```python
sns.lineplot(data=Auto, 
            x = 'year', y = 'displacement',
            hue = 'origin');
```
